```python
# feature selection

#So, id is unique and int type hence can't be a feature
#title does have some reps but we can't exactly treat it as categorical becoz it has 4782 categories!
# we probably can't use the title as feature too
# We can use RATING as a feature (numerical, ordinal)
# MAINCATEGORY : men/ women can also be used as a feature (categorical)
# PLATFORM: amazon/flipkart: feature (categorical)
# ACTPRICE1: feature (numerical,ordinal)
# NORATING1 is the number of ratings of that product i assume: can be used as feature (numerical, ordinal)
# NOREVIEWS1 is the number of reviewss of that product i assume: can be used as feature (numerical, ordinal)
# STAR_5F: no of people who rated it 5 star: feature (numerical, ordinal), similar for other stars
# FULLFILLED is binary (numerical, categorical): feature
```

```python
# NA assignment
traindata2[["star_5f","star_4f","star_3f"]] = traindata2[["star_5f","star_4f","star_3f"]].fillna(0)
traindata2[["star_5f","star_4f","star_3f"]] = traindata2[["star_5f","star_4f","star_3f"]].astype(int)
```

```python
traindata2.head()
```

```python
# new dataframe with only the features we want 
traindata3 = traindata2[["Rating","maincateg","platform","actprice1","norating1","noreviews1","star_5f","star_4f","star_3f","star_2f","star_1f","fulfilled1"]]
```

```python
print(traindata3.isnull().sum())
```

```python
# we can fill the NA in norating1 and noreviews1 with 0 and maincateg with unisex
traindata3[["norating1","noreviews1"]] = traindata3[["norating1","noreviews1"]].fillna(0)
traindata3["maincateg"] = traindata3["maincateg"].fillna("unisex")
print(traindata3.isnull().sum())
```

```python
# all these things also have to be done for the test data

# NA assignment
testdata[["star_5f","star_4f","star_3f","star_1f","Rating"]] = testdata[["star_5f","star_4f","star_3f","star_1f","Rating"]].fillna(0)
testdata[["star_5f","star_4f","star_3f","star_1f","Rating"]] = testdata[["star_5f","star_4f","star_3f","star_1f","Rating"]].astype(int)

# feature selection
# we are directly going from testdata to testdata3 to maintain same count with the training data
testdata3 = testdata[["Rating","maincateg","platform","actprice1","norating1","noreviews1","star_5f","star_4f","star_3f","star_2f","star_1f","fulfilled1"]]

# NA assignment
testdata3[["norating1","noreviews1"]] = testdata3[["norating1","noreviews1"]].fillna(0)
testdata3["maincateg"] = testdata3["maincateg"].fillna("unisex")
```

```python
print(testdata3.isnull().sum())
```

```python
print(targetdata.isnull().sum())
```

```python
# maincateg and platform has to be encoded to 0s and 1s

traindata3 = traindata3.apply(LabelEncoder().fit_transform)
testdata3 = testdata3.apply(LabelEncoder().fit_transform)
```

```python
# train test split for experimentation on traindata3
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(traindata3, targetdata2, test_size = 0.15, random_state = 42)

# check the shape of X_train and X_test
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier, XGBRFClassifier
from xgboost import plot_tree, plot_importance
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
from sklearn import preprocessing
from sklearn.feature_selection import RFE

from sklearn.metrics import mean_squared_error
```

```python
def score(model, title = "Default"):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
#     print(confusion_matrix(y_test, preds))
    rmse = round(mean_squared_error(y_test, preds, squared=False), 5)
    print('RMSE for', title, ':', rmse, '\n')
```
